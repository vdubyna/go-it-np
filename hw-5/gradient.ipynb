```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import time

# Генерація даних
def generate_data(seed=42):
    np.random.seed(seed)
    x1 = np.random.rand(100)
    x2 = np.random.rand(100)

    def polynomial(x1, x2):
        return 4*x1**2 + 5*x2**2 - 2*x1*x2 + 3*x1 - 6*x2

    y = polynomial(x1, x2)
    poly = PolynomialFeatures(degree=2, include_bias=False)
    X_poly = poly.fit_transform(np.column_stack((x1, x2)))
    return X_poly, y

# Функція для побудови графіків
def plot_convergence(errors, method_name, lr, iterations):
    plt.plot(errors, label=f'lr={lr}, iter={iterations}')
    plt.xlabel('Ітерації')
    plt.ylabel('Середньоквадратична помилка (MSE)')
    plt.title(f'Процес збіжності: {method_name}')
    plt.legend()
    plt.grid(True)

# Ліміт збіжності
def check_convergence(errors, tolerance):
    return len(errors) > 1 and abs(errors[-1] - errors[-2]) < tolerance

# Реалізація методів градієнтного спуску
def polynomial_regression_gradient_descent(X, y, lr=0.01, iterations=1000, tolerance=0.0001):
    m, n = X.shape
    theta = np.zeros(n)
    errors = []
    start_time = time.time()

    for i in range(iterations):
        gradients = -2/m * X.T @ (y - X @ theta)
        theta -= lr * gradients
        error = np.mean((y - X @ theta)**2)
        errors.append(error)

        if i % 100 == 0:
            print(f"[Gradient Descent] Iter: {i}, lr: {lr}, Error: {error:.6f}")

        if check_convergence(errors, tolerance):
            print(f"[Gradient Descent] Converged at iteration {i} with error {error:.6f}")
            break

    elapsed_time = time.time() - start_time
    print(f"[Gradient Descent] Total time: {elapsed_time:.4f} seconds")
    return theta, errors

def polynomial_regression_SGD(X, y, lr=0.01, iterations=1000, tolerance=0.0001):
    m, n = X.shape
    theta = np.zeros(n)
    errors = []
    start_time = time.time()

    for i in range(iterations):
        random_index = np.random.randint(m)
        xi = X[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = -2 * xi.T @ (yi - xi @ theta)
        theta -= lr * gradients
        error = np.mean((y - X @ theta)**2)
        errors.append(error)

        if i % 100 == 0:
            print(f"[SGD] Iter: {i}, lr: {lr}, Error: {error:.6f}")

        if check_convergence(errors, tolerance):
            print(f"[SGD] Converged at iteration {i} with error {error:.6f}")
            break

    elapsed_time = time.time() - start_time
    print(f"[SGD] Total time: {elapsed_time:.4f} seconds")
    return theta, errors

def polynomial_regression_rmsprop(X, y, lr=0.01, iterations=1000, beta=0.9, epsilon=1e-8, tolerance=0.0001):
    m, n = X.shape
    theta = np.zeros(n)
    Eg = np.zeros(n)
    errors = []
    start_time = time.time()

    for i in range(iterations):
        gradients = -2/m * X.T @ (y - X @ theta)
        Eg = beta * Eg + (1 - beta) * gradients**2
        theta -= lr * gradients / (np.sqrt(Eg) + epsilon)
        error = np.mean((y - X @ theta)**2)
        errors.append(error)

        if i % 100 == 0:
            print(f"[RMSProp] Iter: {i}, lr: {lr}, Error: {error:.6f}")

        if check_convergence(errors, tolerance):
            print(f"[RMSProp] Converged at iteration {i} with error {error:.6f}")
            break

    elapsed_time = time.time() - start_time
    print(f"[RMSProp] Total time: {elapsed_time:.4f} seconds")
    return theta, errors

def polynomial_regression_adam(X, y, lr=0.01, iterations=1000, beta1=0.9, beta2=0.999, epsilon=1e-8, tolerance=0.0001):
    m, n = X.shape
    theta = np.zeros(n)
    m_t = np.zeros(n)
    v_t = np.zeros(n)
    errors = []
    start_time = time.time()

    for i in range(1, iterations + 1):
        gradients = -2/m * X.T @ (y - X @ theta)
        m_t = beta1 * m_t + (1 - beta1) * gradients
        v_t = beta2 * v_t + (1 - beta2) * gradients**2
        m_t_corr = m_t / (1 - beta1**i)
        v_t_corr = v_t / (1 - beta2**i)
        theta -= lr * m_t_corr / (np.sqrt(v_t_corr) + epsilon)
        error = np.mean((y - X @ theta)**2)
        errors.append(error)

        if i % 100 == 0:
            print(f"[Adam] Iter: {i}, lr: {lr}, Error: {error:.6f}")

        if check_convergence(errors, tolerance):
            print(f"[Adam] Converged at iteration {i} with error {error:.6f}")
            break

    elapsed_time = time.time() - start_time
    print(f"[Adam] Total time: {elapsed_time:.4f} seconds")
    return theta, errors

def polynomial_regression_nadam(X, y, lr=0.01, iterations=1000, beta1=0.9, beta2=0.999, epsilon=1e-8, tolerance=0.0001):
    """
    Nadam (Nesterov-accelerated Adaptive Moment Estimation) для поліноміальної регресії.
    """
    m, n = X.shape
    theta = np.zeros(n)
    m_t = np.zeros(n)  # Середнє значення моментів
    v_t = np.zeros(n)  # Квадратні моменти
    errors = []
    start_time = time.time()

    for i in range(1, iterations + 1):
        gradients = -2/m * X.T @ (y - X @ theta)

        # Оновлення моментів
        m_t = beta1 * m_t + (1 - beta1) * gradients
        v_t = beta2 * v_t + (1 - beta2) * gradients**2

        # Коригування моментів
        m_t_corr = m_t / (1 - beta1**i)
        v_t_corr = v_t / (1 - beta2**i)

        # Оновлення параметрів (Nadam-специфіка)
        nesterov_term = beta1 * m_t_corr + (1 - beta1) * gradients
        theta -= lr * nesterov_term / (np.sqrt(v_t_corr) + epsilon)

        # Розрахунок помилки
        error = np.mean((y - X @ theta)**2)
        errors.append(error)

        if i % 100 == 0:
            print(f"[Nadam] Iter: {i}, lr: {lr}, Error: {error:.6f}")

        # Перевірка на збіжність
        if check_convergence(errors, tolerance):
            print(f"[Nadam] Converged at iteration {i} with error {error:.6f}")
            break

    elapsed_time = time.time() - start_time
    print(f"[Nadam] Total time: {elapsed_time:.4f} seconds")
    return theta, errors

# Порівняння всіх методів
def compare_methods(X, y, learning_rates, iterations_list):
    methods = {
        "Gradient Descent": polynomial_regression_gradient_descent,
        "SGD": polynomial_regression_SGD,
        "RMSProp": polynomial_regression_rmsprop,
        "Adam": polynomial_regression_adam,
        "Nadam": polynomial_regression_nadam
    }

    for method_name, method in methods.items():
        print(f"\n=== {method_name} ===")
        plt.figure(figsize=(10, 6))
        for lr in learning_rates:
            for iter_count in iterations_list:
                _, errors = method(X, y, lr=lr, iterations=iter_count)
                plot_convergence(errors, method_name, lr, iter_count)
        plt.show()

X_poly, y = generate_data()
learning_rates = [0.5, 0.1, 0.01]
iterations_list = [1000]
compare_methods(X_poly, y, learning_rates, iterations_list)
```

* iterations: 1000 - максимальне число ітерацій, робота припинялась при досягненні збіжності
